```python
data_list = []
for word in data:
    if (word[:3]=="관람객"):
        word = word[3:]
    word = word.replace("토니", "Tony")
    word = word.replace("스타크", "Stark")
    word = word.replace("마블", "Marvel")
    word = word.replace("아이언맨", "IronMan")
    word = word.replace("스파이더맨", "SpiderMan")
    word = word.replace("스파이더 맨", "SpiderMan")
    word = word.replace("토르", "Thor")
    word = word.replace("피터", "Peter")
    word = word.replace("파커", "Parker")
    word = word.replace("톰", "Tom")
    word = word.replace("홀랜드", "Holland")
    word = word.replace("해피", "Happy")
    word = word.replace("호건", "Hogan")
    word = word.replace("어벤져스", "Avengers")
    word = word.replace("어벤저스", "Avengers")
    word = word.replace("미스테리오", "Mysterio")
    word = word.replace("코믹스", "comics")
    word = word.replace("소니", "Sony")
    word = word.replace("젠다야", "Zendaya")
    word = word.replace("앤트맨", "Antman")
    word = word.replace("엔드게임", "Endgame")
    word = word.replace("앤드게임", "Endgame")
    data_list.append(word)
print(data_list)
```
