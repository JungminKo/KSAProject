```python
np.random.seed(1234)
x=np.random.binomial(n=100,p=0.5,size=100)
x
```

```python
pd.value_counts(x)
```

```python
cls=[35,40,45,50,55,60]
cls_x=pd.cut(x=x,bins=cls,right=True,include_lowest=True) #includ_lowest=True : 첫번째 하한은 이상으로 취급
cls_x
```

```python
pd.value_counts(cls_x)
```

```python
cls_x=pd.cut(x=x,bins=7,right=True,include_lowest=True)
cls_x
```

```python
pd.value_counts(cls_x)
```

```python
cls_x=pd.cut(x=x,bins=7,labels=["A","B","C","D","E","F","G" ],
             right=True,include_lowest=True)
cls_x
```

```python
pd.value_counts(cls_x)
```

```python
cls_x=pd.qcut(x=x,q=7)
cls_x
```

```python
pd.value_counts(cls_x)
```

```python
cls_x=pd.qcut(x=x,q=5)
cls_x
```
