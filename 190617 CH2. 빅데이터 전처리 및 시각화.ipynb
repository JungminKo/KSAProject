```python
from pandas import Series, DataFrame
from numpy import nan as NA
import pandas as pd
import numpy as np
```

```python
df1=DataFrame({"key":["a","b","a","a","c","b"],"v1":range(6)})
df2=DataFrame({"key":["a","b","d"],"v1":[3,6,4]})
print(df1, "\n", df2)
pd.merge(df1,df2, on="key") # 동일한 변수 기준 결합. 
```

```python
pd.merge(df1,df2) #on을 안쓰면 변수이름똑같은것끼리 결합 
```

```python
# 명칭이 다른 경우
df1=DataFrame({"key1":["a","b","a","a","c","b"],"v1":range(6)})
df2=DataFrame({"key2":["a","b","d"],"v1":[3,6,4]})

pd.merge(df1,df2,left_on="key1",right_on="key2")
```

```python
# 두개 변수 이상 key값
df1=DataFrame({"key1":["a","b","a","a","c","b"],"gender":["M","M","F","F","M","M"],"v1":range(6)})
df2=DataFrame({"key2":["a","b","d"],"sex":["F","M","F"],"v2":[3,6,4]})

print(df1, "\n", df2)
pd.merge(df1,df2,left_on=["key1","gender"],right_on=["key2","sex"])
```

```python
pd.merge(df1,df2,left_on=["key1","gender"],right_on=["key2","sex"],how="left")
```

```python
pd.merge(df1,df2,left_on=["key1","gender"],right_on=["key2","sex"],how="outer")
```

```python
pd.merge(df1,df2,left_on=["key1","gender"],right_on=["key2","sex"],how="outer",sort="True")
```

```python
df1=DataFrame({"key1":["a","b","a","a","c","b"],"gender":["M","M","F","F","M","M"],"v1":range(6)})
df2=DataFrame({"v1":[99,999,9],"v2":[3,6,4]},index=["a","d","b"])

print(df1, "\n", df2)

pd.merge(df1,df2,left_on="key1",right_index=True,how="outer",sort=True)
```

```python
pd.merge(df1,df2,left_index=True, right_index=True,how="outer",sort=True)
```

```python
df1=DataFrame({"gender":["M","M","F","F","M","M"],"v1":range(6)})
df2=DataFrame({"v1":[99,999,9],"v2":[3,6,4]},index=["a","d","b"])

df1.join(df2, how="outer",sort=True, lsuffix="_left",rsuffix="_right")
```
