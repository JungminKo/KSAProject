## 데이터가져오기
### train data

```python
# 파일이름을 넣으면 데이터 불러오기(총 1024개의 값을 불러올 예정)
def digitft(filename):    
    fr=open(filename)
    array0Lines=fr.readlines() #한줄한줄 불러오기
    numberOfLines=len(array0Lines)
    data=[]
    label=[]
    for i in range(numberOfLines):
        array0Lines[i]=array0Lines[i].strip() # \n없애기
        for j in range(len(array0Lines[i])):
            data.append(int(array0Lines[i][j])) # [i,j]에 있는 데이터를 하나씩 정수로 가져온다. 
    label.append(int(filename[26])) # "trainingDigits/"다음에 label이 적혀있으므로 인덱스26을 라벨로 가져온다.
    return data, label
```

```python
# 파일이름 생성하기
list_trn_filename=[]
for i in range(10):
    name_numb=[189,198,195,199,186,187,195,201,180,204] #train data의 갯수는 0부터 9까지 각각 다음과 같다.
    a=int(name_numb[i])
    for j in range(a):
        name='class_data/trainingDigits/'+str(i)+'_'+str(j)+'.txt'
        list_trn_filename.append(name)
```

```python
# 파일이름을 생성한것을 토대로 하나씩 데이터 불러와서 붙이기
trn_data=[]
trn_label=[]
for name in list_trn_filename:
    data, label=digitft(name)
    trn_data=trn_data+[data]
    trn_label=trn_label+label
```

```python
# train data와 동일한 방법으로 불러오기. 단, label을 붙일 때 파일이름이(위치가) 바뀌었으므로 인덱스22를 넣어줘야한다
def digitft_1(filename):    
    fr=open(filename)
    array0Lines=fr.readlines()
    numberOfLines=len(array0Lines)
    data=[]
    label=[]
    for i in range(numberOfLines):
        array0Lines[i]=array0Lines[i].strip()
        for j in range(len(array0Lines[i])):
            data.append(int(array0Lines[i][j]))
    label.append(int(filename[22])) #이부분만 train data와 다르다
    return data, label
```

## 1. K-NN사용
the total number of errors is: 11.0

the total error rate is:0.011627906976744186

```python
def classify0(inX, dataSet, labels, k): #K-NN함수 가져오기
    dataSetSize = dataSet.shape[0]
    diffMat = np.tile(inX, (dataSetSize, 1)) - dataSet
    sqDiffMat = diffMat ** 2
    sqDistances = sqDiffMat.sum(axis = 1)
    distances = sqDistances ** 0.5
    sortedDistIndicies = distances.argsort()
    classCount={}
    for i in range(k):
        voteIlabel = labels[sortedDistIndicies[i]]
        classCount[voteIlabel] = classCount.get(voteIlabel, 0) + 1
    sortedClassCount = sorted(classCount.items(),
                              key=operator.itemgetter(1), reverse=True)
    return sortedClassCount[0][0]
```

```python
def ClassTest(): #test data를 가져와서 맞는지 확인해보기
    trn_dat=np.array(trn_data)
    tst_dat=np.array(tst_data)
    numTestVecs=tst_dat.shape[0]
    errorCount=0.0
    for i in range(numTestVecs):
        classifierResult=classify0(tst_dat[i,:],trn_dat,trn_label,3)
        print("the classifier came back with: {}, the real answer is: {}".format(classifierResult,tst_label[i]))
        if(classifierResult != tst_label[i]):errorCount+=1.0
    print("the total number of errors is: {}".format(errorCount))
    print("the total error rate is:{}".format(errorCount/float(numTestVecs)))
```

## 2. Naive Bayes
the total number of errors is: 71.0

the total error rate is:0.07505285412262157

```python
def trainNBO(trainMatrix,trainCategory):
    numTrainMat=len(trainMatrix)
    numFeat=len(trainMatrix[0])
    pLabel=[] #라벨이 나오는 확률을 각각 계산해서 list로 만들기
    pVect=[] #라벨별로 각 자리에 1이 나올 확률을 계산해서 vector로 만들려고한다!!!
    for i in range(10):
        pLab=float(trn_label.count(i))/float(numTrainMat)
        pLabel.append(pLab) # 0부터 9까지 라벨이 나오는 확률을 각각 계산해서 list로 만듦
        pNum=np.ones(numFeat) 
        pDenom=2.0
        for j in range(numTrainMat):    
            if trainCategory[j]==i:
                pNum+=trainMatrix[j]
                pDenom+=sum(trainMatrix[j])
        pVec=np.log(pNum/pDenom)
        pVect=pVect+[pVec]
    return pVect,pLabel
```

```python
def classifyNB(testVect,pVect,pLabel):
    testVect=np.array(testVect)
    P=[]
    for i in range(10): #testVect일 때 각 라벨이 될 확률 계산하기
        p=sum(testVect*pVect[i])+np.log(pLabel[i])
        P.append(p)
    P=pd.Series(P)
    return P.idxmax() #라벨별로 계산한후 가장 큰 라벨 뽑아내기
```

```python
def testingNB():
    
    errorCount=0.0
    n=np.array(tst_data).shape[0]
    pVect,pLabel=trainNBO(trn_data,trn_label)
    
    for i in range(n):
        NBResult = classifyNB(tst_data[i], pVect, pLabel)
        print("the classifier came back with: {}, the real answer is: {}".format(NBResult,tst_label[i]))
        if(NBResult != tst_label[i]):errorCount+=1.0
    print("the total number of errors is: {}".format(errorCount))
    print("the total error rate is:{}".format(errorCount/float(n)))
```

```python
from numpy import *
from os import listdir
import operator
```

```python
def classify0(inX, dataSet, labels, k):
    dataSetSize = dataSet.shape[0]
    diffMat = tile(inX, (dataSetSize, 1)) - dataSet
    sqDiffMat = diffMat ** 2
    sqDistances = sqDiffMat.sum(axis = 1)
    distances = sqDistances ** 0.5
    sortedDistIndicies = distances.argsort()
    classCount={}
    for i in range(k):
        voteIlabel = labels[sortedDistIndicies[i]]
        classCount[voteIlabel] = classCount.get(voteIlabel, 0) + 1
    sortedClassCount = sorted(classCount.items(),
                              key=operator.itemgetter(1), reverse=True)
    return sortedClassCount[0][0]
```

```python
def img2vector(filename):
    returnVect = zeros((1, 1024))
    fr = open(filename)
    for i in range(32):
        lineStr = fr.readline()
        for j in range(32):
            returnVect[0, 32*i+j] = int(lineStr[j])
    return returnVect
```

```python
def DigitClassifier():
    trnLabels = []
    trainingFileList = listdir('class_data/trainingDigits')
    m = len(trainingFileList)
    trainingMat = zeros((m, 1024))
    for i in range(m):
        fileNameStr = trainingFileList[i]
        fileStr = fileNameStr.split('.')[0]
        classNumStr = int(fileStr.split('_')[0])
        trnLabels.append(classNumStr)
        trainingMat[i, :] = img2vector('class_data/trainingDigits/%s' % fileNameStr)
        
    tstLabels = []  
    tstFileList = listdir('class_data/testDigits')
    n = len(tstFileList)
    tstMat = zeros((n, 1024))
    for i in range(n):
        fileNameStr = tstFileList[i]
        fileStr = fileNameStr.split('.')[0]
        classNumStr = int(fileStr.split('_')[0])
        tstLabels.append(classNumStr)
        tstMat[i, :] = img2vector('class_data/testDigits/%s' % fileNameStr)

    errorCount=0.0
    for i in range(n):
        classifierResult = classify0(tstMat[i,:], trainingMat, trnLabels, 3)
        print("the classifier came back with: {}, the real answer is: {}".format(classifierResult,tstLabels[i]))
        if(classifierResult != tstLabels[i]):errorCount+=1.0
    print("the total number of errors is: {}".format(errorCount))
    print("the total error rate is:{}".format(errorCount/float(n)))
```
